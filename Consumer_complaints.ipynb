```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Consumer_Complaints.csv (3).zip')
```

```python
df.head()
```

```python
df.dropna(subset=['Consumer complaint narrative', 'Product'])
```

```python
df['Date received'] = pd.to_datetime(df['Date received'])
```

```python
top_products = df['Product'].value_counts().head(5)
print("Top 5 complaint Products:")
print(top_products)
```

```python
top_products.plot(kind='bar', title='Top 5 Complaint categories')
plt.ylabel('Number of Complaints')
plt.show()
```

```python
if 'Date sent to company' in df.columns:
  df ['response time'] = (pd.to_datetime(df['Date sent to company']) - df['Date received']).dt.days
print("Average response time:", df['response time'].mean())
```
